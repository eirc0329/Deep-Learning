```python
# inline plotting instead of popping out
%matplotlib inline

import os, math
from IPython.display import Image
import numpy as np
seed = np.random.seed

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
class AdalineSGD(object):
    """ADAptive LInear NEuron classifier.

    Parameters
    ------------
    eta : float
        Learning rate (between 0.0 and 1.0)
    n_iter : int
        Passes over the training dataset.

    Attributes
    -----------
    w_ : 1d-array
        Weights after fitting.
    errors_ : list
        Number of misclassifications in every epoch.
    shuffle : bool (default: True)
        Shuffles training data every epoch if True to prevent cycles.
    random_state : int
        Set random state for shuffling and initializing the weights.
        
    """
    def __init__(self, eta=0.01, n_iter=50, shuffle=True, random_state=1, M=1):
        self.eta = eta
        self.n_iter = n_iter
        self.w_initialized = False
        self.shuffle = shuffle
        if random_state : np.random.seed(random_state)
        self.batch_size = M
        
    def fit(self, X, y):
        """ Fit training data.

        Parameters
        ----------
        X : array-like; shape = [n_samples, n_features]
            Training vectors, where n_samples is the number of samples and
            n_features is the number of features.
        y : array-like; shape = [n_samples]
            Target values or labels.

        Returns
        -------
        self : object

        """
        self._initialize_weights(X.shape[1])
        self.cost_ = []

        for i in range(self.n_iter):
            if self.shuffle:
                X, y = self._shuffle(X, y)
            cost = []
            for b in range(int(len(y)/self.batch_size)):
                X_b = X[self.batch_size*b:self.batch_size*(b+1)]
                y_b = y[self.batch_size*b:self.batch_size*(b+1)]
                cost.append(self._update_weights(X_b, y_b))                            
            avg_cost = sum(cost) / len(y)
            self.cost_.append(avg_cost)
        return self

    def _shuffle(self, X, y):
        """Shuffle training data"""
        r = np.random.permutation(len(y))
        return X[r], y[r]
    
    def _initialize_weights(self, m):
        """Randomly initialize weights"""
        self.w_ = np.random.normal(loc=0.0, scale=0.01, size=1 + m)
        self.w_initialized = True
        
    def _update_weights(self, xi, yi):
        """Apply Adaline learning rule to update the weights"""

        #output is a multi-dimension
        output = []
        for x in xi:
          output.append(self.activation(x))

        # Cost function
        cost = 0
        for output, y in zip(output, yi):
          error = (y - output)
          cost += 0.5 * error**2
          
          # Update rule
          self.w_[1:] += self.eta * x.dot(error)
          self.w_[0] += self.eta * error


        return cost
    
    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        """Compute linear activation"""
        return self.net_input(X)

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.activation(X) >= 0.0, 1, -1)

    def partial_fit(self, X, y):
        """Fit training data without reinitializing the weights"""
        if not self.w_initialized:
            self._initialize_weights(X.shape[1])                
        if y.ravel().shape[0] > 1:
            for xi, yi in zip(X, y):
                self._update_weights(xi, yi)
        else:
            self._update_weights(X, y)
        return self
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from IPython.display import display

df = pd.read_csv('https://archive.ics.uci.edu/ml/'
        'machine-learning-databases/iris/iris.data', header=None)
df.columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class label']
display(df.head())

X = df[['Petal length', 'Petal width']].values
y = pd.factorize(df['Class label'])[0]

# discard exmaples in the first class
X = X[50:150]
y = np.where(y[50:150] == 2, -1, y[50:150])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

print('# Training data points: %d' % X_train.shape[0])
print('# Testing data points: %d' % X_test.shape[0])
print('Class labels: %s' % np.unique(y))
```

```python
ada1 = AdalineSGD(n_iter=20, eta=0.01, M=1, random_state=1).fit(X_train_std, y_train)
ada2 = AdalineSGD(n_iter=20, eta=0.01, M=5, random_state=1).fit(X_train_std, y_train)
ada3 = AdalineSGD(n_iter=20, eta=0.01, M=10, random_state=1).fit(X_train_std, y_train)

# Testing accuracy
y_pred1 = ada1.predict(X_test_std)
print("Batch size = 1")
print('Misclassified samples: %d' % (y_test != y_pred1).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred1))

y_pred2 = ada2.predict(X_test_std)
print("\nBatch size = 5")
print('Misclassified samples: %d' % (y_test != y_pred2).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred2))

y_pred3 = ada3.predict(X_test_std)
print("\nBatch size = 10")
print('Misclassified samples: %d' % (y_test != y_pred3).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred3))


plt.plot(range(1, len(ada1.cost_)+1), ada1.cost_, marker='o', label='M=1', )
plt.plot(range(1, len(ada2.cost_)+1), ada2.cost_, marker='s', label='M=5')
plt.plot(range(1, len(ada3.cost_)+1), ada3.cost_, marker='x', label='M=10')
plt.xlabel('Epochs')
plt.ylabel('Sum-squared-error')
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('./output/Lab04-1_112062559.png', dpi=300)
plt.show()
```
